## Setup

Paste API key as Colab secret: `GEMINI_API_KEY`

Steps:
1. Open the **Secrets** panel (key icon in left sidebar).
2. Add a secret named `GEMINI_API_KEY` with your Google AI Studio API key as the value.
3. Enable notebook access for that secret.
4. Run Cell 2 to install dependencies and configure the model.

```python
!pip install -q google-generativeai
import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-2.0-pro-exp')
print('Model ready:', model.model_name)
```

## Delegate

Paste Claude prompt below.

1. Replace the placeholder text inside the triple-quoted string in Cell 4.
2. Run Cell 4.
3. The response is printed and saved to `/content/gemini_response.txt`.
4. Copy the output back into your Claude Code session.

```python
prompt = """PASTE HERE"""
response = model.generate_content(prompt)
print(response.text)
with open('/content/gemini_response.txt', 'w') as f: f.write(response.text)
print("Saved! Copy to Claude.")
```
